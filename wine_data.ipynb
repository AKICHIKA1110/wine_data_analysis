```python
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np

# 
red_wine_path = './WineQuality-RedWine.csv'
white_wine_path = './WineQuality-WhiteWine.csv'

# 
red_wine = pd.read_csv(red_wine_path)
white_wine = pd.read_csv(white_wine_path)

# 
plt.figure(figsize=(10, 6))
sns.countplot(x='quality', data=red_wine)
plt.title('Red Wine Quality Distribution')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.savefig('./images/red_wine_quality_distribution.png')
plt.show()

# 
plt.figure(figsize=(10, 6))
sns.countplot(x='quality', data=white_wine)
plt.title('White Wine Quality Distribution')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.savefig('./images/white_wine_quality_distribution.png')
plt.show()

# 
plt.figure(figsize=(12, 10))
sns.heatmap(red_wine.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Red Wine Feature Correlation')
plt.savefig('./images/red_wine_feature_correlation.png')
plt.show()

# 
plt.figure(figsize=(12, 10))
sns.heatmap(white_wine.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('White Wine Feature Correlation')
plt.savefig('./images/white_wine_feature_correlation.png')
plt.show()
```

```python
import pandas as pd
import numpy as np
import plotly.express as px
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# 
red_wine_path = './WineQuality-RedWine.csv'

# 
red_wine = pd.read_csv(red_wine_path)

# 
X = red_wine.drop('quality', axis=1)  # 
y = red_wine['quality']  # 

# 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 
df = pd.DataFrame(X_scaled, columns=X.columns)
df['quality'] = y

# 
df_melted = pd.melt(df, value_vars=X.columns, id_vars=['quality'])

# 
fig = px.scatter(df_melted, x="value", y="quality", color="variable", facet_col="variable", facet_col_wrap=2, labels={"quality": "Quality", "value": "Standardized Value"})

# x
for axis in fig.layout:
    if axis.startswith('xaxis') or axis.startswith('yaxis'):
        fig.layout[axis].showticklabels = True

# 
fig.update_layout(
    margin=dict(l=40, r=40, t=40, b=40),
    grid={'rows': 3, 'columns': 3, 'pattern': 'independent'},
    height=1200,
    width=800,
    title='Red Wine scatter'
)

# 

fig.show()
```
